```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'species']

df = pd.read_csv(url, names=names)

df
```

```python
# 表格的形狀（row,column）
print(df.shape)

# 取出某個欄位
print(df["sepal-length"].mean())
print(df["sepal-length"].std())
print(df["sepal-length"])
```

```python
# 用條件取出表中的值
# df[True or False]
df[df["species"]=="Iris-setosa"]
```

```python
df.query("species == 'Iris-setosa'")
```

```python
# 某個欄位相異的值

df['species'].unique()
```

```python
# 將某個欄位相同的人 group 起來

df.groupby('species').size()
```

```python
# 將某個欄位相同的人 group 起來

df.groupby('species').mean()
```

```python
%matplotlib inline

ax = df[df["species"]=="Iris-setosa"].plot.scatter(x="sepal-length",y="sepal-width",c="red",label="Iris-setosa")
ax = df[df["species"]=="Iris-versicolor"].plot.scatter(x="sepal-length",y="sepal-width",c="blue",label="Iris-versicolor",ax=ax)
ax = df[df["species"]=="Iris-virginica"].plot.scatter(x="sepal-length",y="sepal-width",c="green",label="Iris-virginica",ax=ax,figsize=(10,10))
```

```python
df.boxplot(["sepal-length","sepal-width"],by="species",figsize=(10,10))
```

```python
# 新增一個欄位

df["length"] = df["sepal-length"] + df["petal-length"]
df
```

```python
# 對某個欄位做 standard normorlization

df["std"] = (df["sepal-length"] - df["sepal-length"].mean()) / df["sepal-length"].std()
df
```

```python
import pandas as pd
df = pd.read_csv('/Users/hank/py/2019-spring/gossip0506.csv')
df
```

```python
df.plot.line(y="all",figsize=(10,6))
```

```python
ax = df.plot.line(y="all",figsize=(10,6),c='blue')
ax = df.plot.line(y="up",figsize=(10,6),c='green',ax=ax)
ax = df.plot.line(y="down",figsize=(10,6),c='red',ax=ax)
```

```python
import numpy as np
df.plot.bar(figsize=(10,6))
```

```python
import matplotlib.pyplot as plt

time_label = ['{:02d}'.format(i//6) for i in range(144)]
plt.figure(figsize=(20,10))

plt.bar(np.arange(144), df['all'], align='center', alpha=0.5)

plt.xticks(np.arange(144,step=6), time_label[::6])
plt.ylabel('Comment')
plt.title('PTT Gossip 2/24')
 
plt.show()
```
