## While Loop

- 當條件成立，就執行 block 裡面的程式碼，之後再回到while開頭再問一次條件是否成立
- 條件不成立，就不執行 block 裡面的程式碼
- while 的屁股要加上 `:` （冒號）
- 記得用縮排 `indent` 當作 block

```python
while <condition>:
    <statement>
    <statement>
```

![](https://upload.wikimedia.org/wikipedia/commons/thumb/4/43/While-loop-diagram.svg/220px-While-loop-diagram.svg.png)

````python
x = 10
print('begin of while loop')

while x >= 0:
    print('x > 0')
    
print('end of while loop')
````

````python
x = 10

while x >= 0:
    print('x > 0',x)
    x -= 1  # x = x - 1
print('end of while loop')
````

## `break`

- 執行到 break 就跳出迴圈

## `continue`

- 直接跳回迴圈的開頭（條件判斷）

````python
x = 10
while x > 0:
    x -= 1
    if x == 6:
        break
    print(x)
    
````

````python
x = 10
while x > 0:
    x -= 1
    if x > 3 and x < 7:
        continue  
    print(x)
    
````

````python
while True:
    inp = input()
    print('hello,',inp)
    if inp == '0':
        break
````

# for loop

- 遍歷整個 sequence ，每次 loop 拿出一個東西
- 也一樣要用 `indent` 排版出 `block`
- 後面要加上 `:`(冒號)
- 和 `while` 一樣可以 `break` 和 `continue`

```python
for <variable> in <sequence>:
    <statement>
    <statement>
```

````python
fruits = ['apple','orange','mango','banana']

for f in fruits:
    print(f)
````

## `range()`

`range(start, stop[, step])`

- 跟 slice 的用法很像：

|用法|效果|
|-|-|
| `range(a)` | 0 ~ a-1|
| `range(a,b)` | a ~ b-1 |
| `range(a,b,c)` |a ~ b-1 ，每次間隔c|

````python
for i in range(5):
    print(i)
    
print('-'*20)
for i in range(5,10):
    print(i)
    
print('-'*20)
for i in range(2,11,2):
    print(i)
````

````python
x = 'hello world!'
for letter in x:
    print(letter)
````

## `zip(x,y)`

- 兩種物件一起迭代
- 次數以比較短的為主

````python
fruits = ['apple','orange','mango','banana']

for i, f in enumerate(fruits):
    print(i,f)
````

````python
people = ['hank','jason','alice','bob','sandy']
fruits = ['apple','orange','mango','banana']

for p, f in zip(people,fruits):
    print(p,'loves',f)
````
