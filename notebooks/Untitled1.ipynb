```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'species']

df = pd.read_csv(url, names=names)

df
```

```python
df["petal-width"].mean()
```

```python
df.dtypes
```

```python
df_setosa = df[df["species"]=="Iris-setosa"]
df_setosa["petal-width"].mean()
```

```python
df['sepal-length'] = df['sepal-width'].astype(float)
df[ df['sepal-length'] >= 5.]
```

```python
df.groupby('species').std()
df.groupby('species').mean()
```

```python
df.iloc[0:3,1:3]
```

```python
%matplotlib inline

ax = df[df["species"]=="Iris-setosa"].plot.scatter(x="sepal-length",y="sepal-width",c="red",label="Iris-setosa")
ax = df[df["species"]=="Iris-versicolor"].plot.scatter(x="sepal-length",y="sepal-width",c="blue",label="Iris-versicolor",ax=ax,figsize=(10,10))
ax = df[df["species"]=="Iris-virginica"].plot.scatter(x="sepal-length",y="sepal-width",c="green",label="Iris-virginica",ax=ax)

```

```python
df.boxplot(["sepal-length","sepal-width"],by="species",figsize=(10,10))
```
