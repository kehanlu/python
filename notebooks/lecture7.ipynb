## 案例：

每一篇文章的推噓文數

````python
import requests
from bs4 import BeautifulSoup

resp = requests.get("https://www.ptt.cc/bbs/Soft_Job/index.html")
bs = BeautifulSoup(resp.text, 'html.parser')
bs
````

````python
bs.find_all("a")
````

````python
# 所有的標題超連結
posts_title = bs.find_all("div","title")
for post in posts_title:

    # 有些文章被刪除了，只拜訪沒有被刪除的
    if post.a:    
        print(post.text.strip())
        
        # 文章標題
        post_url = post.a['href'].strip()
        print(post_url)
        
        # 拜訪該文章頁面
        post_resp = requests.get("https://www.ptt.cc/" + post_url)
        post_bs = BeautifulSoup(post_resp.text, 'html.parser')
        
        # 抓出推噓箭頭
        push = post_bs.find_all("div","push")
        print("推噓箭頭：", len(push))
        print()
````

## HTTP request 

（常見）

`GET`

GET 方法請求展示指定資源。使用 GET 的請求只應用於取得資料。

`POST`

POST 方法用於提交指定資源的實體，通常會改變伺服器的狀態或副作用（side effect）。

### GET

GET 方法請求展示指定資源。使用 GET 的請求只應用於取得資料。

`URL` + `querystring`(key=value)

ex:

`example.com?a=apple&b=banana`

`https://www.google.com/search?&q=python`

### POST

POST 方法用於提交指定資源的實體，通常會改變伺服器的狀態或副作用（side effect）。

通常用在表單、送資料給伺服器⋯⋯

````python
import requests

payload = {'key1': 'value1', 'key2': 'value2'}
r = requests.get("http://httpbin.org/get", params=payload)
print(r.text)

print("=="*20)

payload = (('key1', 'value1'), ('key1', 'value2'))
r = requests.post('http://httpbin.org/post', data=payload)
print(r.text)
````

## 和我一起修課的同學們

`requests.session`

````python
import requests
from getpass import getpass
from bs4 import BeautifulSoup


session = requests.Session() # enable session


# login

LOGIN_URL = "https://moodle.ntust.edu.tw/login/"

login_token = BeautifulSoup(session.get(LOGIN_URL).text,'html.parser').find("input",attrs={"name":"logintoken"})["value"]

login_data = {
    'username': input(),
    'password': getpass(),
    'logintoken': login_token,
    'rememberusername': 1
}

resp = session.post(LOGIN_URL, login_data)
print("login...\n\n")
print(resp.text)
````

````python
def mask(name):
    if name.split()[0] == "老師@":
        return "老師@" + name.split()[1][:1] + "XX"
    else:
        return name.split()[0][:-4] + "@ " + name.split()[1][:1] + "XX"
````

````python
# parse

bs = BeautifulSoup(resp.text, 'html.parser')

for course in bs.find_all("h3", "coursename"):
    print('\n\n',course.a.text)
    course_id = course.a['href'].split("=")[-1]
    print(course_id)
    
    # 爬上課的同學
    classmate_resp = session.get("https://moodle.ntust.edu.tw/user/index.php?id={}&perpage=1000".format(course_id))
    classmates = BeautifulSoup(classmate_resp.text,'html.parser').find("table",id="participants").find_all("tr")
    for classmate in classmates[1:]:
        name = classmate.find("td","cell c1").text
        if not name:
            break
        print(mask(name))
        
        
    
````
