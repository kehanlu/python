## å’Œæˆ‘ä¸€èµ·ä¿®èª²çš„åŒå­¸å€‘

`requests.session`

````python
import requests
from getpass import getpass
from bs4 import BeautifulSoup


session = requests.Session() # enable session


# login

LOGIN_URL = "https://moodle.ntust.edu.tw/login/"

login_token = BeautifulSoup(session.get(LOGIN_URL).text,'html.parser').find("input",attrs={"name":"logintoken"})["value"]

login_data = {
    'username': input(),
    'password': getpass(),
    'logintoken': login_token,
    'rememberusername': 1
}

resp = session.post(LOGIN_URL, login_data)
print("login...\n\n")
print(resp.text)
````

````python
def mask(name):
    if name.split()[0] == "è€å¸«@":
        return "è€å¸«@" + name.split()[1][:1] + "XX"
    else:
        return name.split()[0][:-4] + "@ " + name.split()[1][:1] + "XX"
````

````python
# parse

bs = BeautifulSoup(resp.text, 'html.parser')

for course in bs.find_all("h3", "coursename"):
    print('\n\n',course.a.text)
    course_id = course.a['href'].split("=")[-1]
    print(course_id)
    
    # çˆ¬ä¸Šèª²çš„åŒå­¸
    classmate_resp = session.get("https://moodle.ntust.edu.tw/user/index.php?id={}&perpage=1000".format(course_id))
    classmates = BeautifulSoup(classmate_resp.text,'html.parser').find("table",id="participants").find_all("tr")
    for classmate in classmates[1:]:
        name = classmate.find("td","cell c1").text
        if not name:
            break
        print(mask(name))
        
        
    
````
