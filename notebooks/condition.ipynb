## `if statement`

- `if` 第一個條件
- `elif` 其他條件，可以有 0~多個
- `else` 沒有符合前述 condition 的其他狀況，可以不使用
- 每一組只會有一個（或沒有）條件成立
- 記得在屁股加上 `:` (冒號)
- 需要以一個縮排`indent`（實際上為四個空白），做「排版」

```python
if <condition>:
    <statement>
    <statement>
elif <condition>:
    <statement>
else:
    <statement>
```

````python
x = 5

if x > 5:
    print("is > 5")
elif x > 2:
    print("is > 2")
elif x > 0:
    print("is > 0")
else:
    print("is <= 2")
````

## 比較複雜的邏輯

|運算子|作用|
|--|--|
|`not x`|反向邏輯（True to False,False to True）|
|`a and b`||
|`a or b`||
|`a in b`|a 屬於 b|

````python
print(not True)
print(not False)

print(True and True and False) # and: 其中有 False 就是 False

print(True or False or False) # or: 其中有 True 就是 True
````

## 巢狀結構(nest)

- 很多層
- 只要記清楚 if 的規則，就可以很輕鬆的判斷
- 不同縮排（`indent`），為不同的 if statement

````python
x = 10
if x > 5:
    print("x>5")
    
    if x > 10:
        print("x>10")
    else:
        print("x>5 but x<=10")
else:
    print("x<=5")
````
